# Mathematical definition of probability

Probability is a measure of uncertainty. As a mathematical theory, it assigns a number in $[0, 1]$ to events. Events are outcomes of experiments. The set of all outcomes is called the sample space $\Omega$. An event is a subset of $\Omega$. If $\Omega$ is finite then every subset is an event. Otherwise, only those which form a $\sigma$-algebra are events. A $\sigma$-algebra over a set $\Omega$ is a family of subsets $\mathcal{F}$ of $\Omega$ such that:

- $\mathcal{F}$ is non-empty.
- If $A \in \mathcal{F}$ then $A^c = \Omega - A \in \mathcal{F}$.
- If $A_1, A_2, \ldots$ is a countable collection of subsets of $\Omega$ then so is $\bigcup_{k \ge 1}A_k$.

A probability measure or a probability distribution is an assignment of a probability to each member of $\mathcal{F}$ (If $\Omega$ is finite then we can choose $\mathcal{F} = 2^\Omega$.) subject to the following conditions:

- $P(A) \in [0, 1]$ for all $A \in \mathcal{F}$.
- $P(A^c) = 1 - P(A)$.
- If $A_1, A_2, \ldots$ are pairwise disjoint members of $\mathcal{F}$ then
$$
P\left(\bigcup_{k \ge 1} A_k\right) = \sum_{k \ge 1} P(A_k).
$$

These three requirements are called as Kolmogorov's axioms.

# Interpretation of probability

The probability of an event can be interpreted in at least two different ways:

- A measure of our degree of belief in the event happening.
- An empirical measure of the ratio of the event happening in a large number of repititions of an experiment.

Irrespective of the interpretation, the probability measure must satisfy the Kolmogorov's axioms.

# Independent events

Events $A$ and $B$ are said to be independent if $P(AB) = P(A)P(B)$, where the event $AB = A \cap B$. Independence of events is either assumed or inferred. An example of the former is when we assume that successive tosses of the same coin are independent of each other. An example of the latter is when we calculate the three probabilities $P(A), P(B), P(AB)$ and infer the relation $P(AB) = P(A)P(B)$.

Independence of events is in some ways quite opposite of the events being disjoint or mutually exclusive. If $A$ and $B$ are mutually exclusive then $P(AB) = 0$ although neither $P(A)$ nor $P(B)$ may be zero.

# Conditional probability

The conditional probability of an event $A$ given an event $B$ is denoted by $P(A|B)$. It is seen to be $P(AB)/P(B)$, where we assume that $P(B) > 0$. If $A$ and $B$ are independent then $P(A|B) = P(A)$. This fact further emphasises the meaning of independence that knowledge of one of them occurring has no impact on the probability of the other one happening.

It follows that
$$
P(B|A) = \frac{P(AB)}{P(A)} ⇒ P(AB) = P(B|A)P(A)
$$
so that
$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)} = \left(\frac{P(B|A)}{P(B)}\right)P(A).
$$
$P(A)$ is called the prior probability while $P(A|B)$ is called the posterior probability of $A$. The posterior probability if our updated degree of belief after getting to know that $B$ has happened. The term $P(B|A)/P(B)$ is called the **likelihood ratio**.


# A few lemmas

**Lemma 1**: If $A$ and $B$ are events then $P(A \cup B) = P(A) + P(B) - P(A \cap B)$.

**Proof**: $A \cup B = (A \cap B^c) \cup (A \cap B) \cup (A^c \cap B)$. The three events whose union is the right hand side on the equation are disjoint. Therefore,
\begin{eqnarray}
P(A \cup B) &=& P(A \cap B^c) + P(A \cap B) + P(A^c \cap B) \\
 &=& P(A \cap B^c) + P(A \cap B) + P(A^c \cap B) + P(A \cap B) - P(A \cap B) \\
 &=& P(A \cap (B^c \cup B)) + P((A^c \cup A) \cap B) - P (A \cap B) \\
 &=& P(A) + P(B) - P(A \cap B)
\end{eqnarray}
$\square$

**Lemma 2**: Let $A_1, A_2, \ldots$ be a monotone increasing sequence of events such that $A_n \rightarrow A$. Then
$$
\lim_{n \rightarrow \infty}P(A_n) = P(A).
$$
**Proof**: Given that $A_1 ⊆ A_2 ⊆ \ldots$. Define $B_1 = ∅$ and $B_n = A_n - A_{n-1}$. Then $A_n = \cup_{k \ge 1}^nA_k = \cup_{k \ge 1}^nB_k$. The events $B_k$ are pair-wise disjoint. Therefore,
$$
P(A_n) = \sum_{k=1}^n P(B_k)
$$
and
$$
\lim_{n \rightarrow \infty}P(A_n) = \lim_{n \rightarrow \infty}\sum_{k=1}^n P(B_k) = \lim_{n \rightarrow \infty}P\left(\cup_{k=1}^n B_k\right) = P(A).
$$
$\square$

**Lemma 3**: Let $A_1, A_2, \ldots$ be a monotone decreasing sequence of events such that $A_n \rightarrow A$. Then
$$
\lim_{n \rightarrow \infty}P(A_n) = P(A).
$$
**Proof**: Given that $A_1 ⊇ A_2 ⊇ \ldots$ so that $A_1^c ⊆ A_2^c ⊆ \ldots$.

$A_n \rightarrow A$ means that $\cap_{k \ge 1}A_k = A$. This also means that $A_n^c \rightarrow A^c$ and $\cup_{k \ge 1}A_k^c = A^c$. Using the previous lemma, we have $P(A_n^c) \rightarrow P(A^c)$ so that $1 - P(A_n) \rightarrow 1 - P(A)$.

$\square$

**Theorem 1**: (Law of total probability) If $A_1, \ldots, A_n$ partition the sample space $\Omega$ then
$$
P(B) = \sum_{k=1}^nP(B|A_k)P(A_k).
$$
**Proof**: Let $C_k = B \cap A_k$ then the events $C_k$ are disjoint and $\cup_{k=1}^n C_k = B$. Therefore,
$$
P(B) = \sum_{k=1}^nP(C_k) = \sum_{k=1}^nP(B \cap A_k) = \sum_{k=1}^nP(B|A_k)P(A_k).
$$
$\square$

**Theorem 2**: (Bayes) Let $A_1, \ldots, A_n$ partition $\Omega$ and $B$ be an event with non-zero probability. Then
$$
P(A_k|B) = \frac{P(B|A_k)P(A_k)}{\sum_{k=1}^nP(B|A_k)P(A_k)}
$$

**Proof**: Follows immediately from the definition of conditional probability and theorem 1.

$\square$

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 21

```python
def decide(x: float, p: float) -> int:
    if x < p:
        return 1
    else:
        return 0

def simulate(p: float, N: int) -> float:
    coin_flips = [decide(x, p) for x in np.random.uniform(0, 1, N)]
    return sum(coin_flips)/N

N = 1000
Ps = np.linspace(start=3, stop=90, num=30)/100
p_hat = [simulate(x, N) for x in Ps]
df = pd.DataFrame({'p': Ps, 'phat': p_hat})
```

```python
plt.plot([Ps[0], Ps[0]], [Ps[len(Ps) - 1], Ps[len(Ps) - 1]])
ax = sns.scatterplot(df, x='p', y='phat')
sns.regplot(df, x='p', y='phat', line_kws={'color': 'g'})
ax.set(xlabel = 'p', ylabel = 'estimated p')
_ = ax.set_title('Estimated vs parametric probability')
```

# Problem 23

```python
def decideB(n: int) -> int:
    if n >= 1 and n <= 4:
        return 1
    else:
        return 0

def decideA(n: int) -> int:
    if n % 2 == 0:
        return 1
    else:
        return 0

def decideAB(n: int) -> int:
    if n == 2 or n == 4:
        return 1
    else:
        return 0

def simulate_die_rolls(n_rolls: int) -> float:
    die_rolls = [np.random.choice(np.arange(1, 7), 1)[0].item() for n in range(n_rolls)]
    pA = sum([decideA(r) for r in die_rolls])/n_rolls
    pB = sum([decideB(r) for r in die_rolls])/n_rolls
    pAB = sum([decideAB(r) for r in die_rolls])/n_rolls

    return (pA * pB)/pAB

n_repetitions = 100
results = [simulate_die_rolls(1000) for n in range(n_repetitions)]
```

The list `results` should ideally have all ones because the two events $A$ and $B$ are independent. However, in a simulation, they deviate from their expected value. The histogram in the next cell shows their distribution.

```python
ax = sns.histplot(results, bins=20, kde=True)
_ = ax.set_title('Ratio of estimated probabilities')
```
